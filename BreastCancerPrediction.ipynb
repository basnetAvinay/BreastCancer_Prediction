```python
import pandas as pd
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
from sklearn.linear_model import LogisticRegression 
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('breast-cancer.csv')
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
y = df['Class']
```

```python
x = df[[ 'menopause', 'Deg-malig','irradiat','Inv-nodes','Tumor-size','Node-caps','Breast-quad']]
```

```python
x
```

```python
y
```

```python
x = pd.get_dummies(x, columns=[ 'menopause', 'Deg-malig','irradiat','Inv-nodes','Tumor-size','Node-caps','Breast-quad'])
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.score(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
y_pred = lr.predict(x_test)
```

```python
for idx, y in enumerate(y_test):
    print('Real:' + str(y) + ' Predicted:'+str(y_pred[idx]))
```

```python
classfier = DecisionTreeClassifier()
```

```python
classfier.fit(x_train, y_train)
```

```python
classfier.score(x_train, y_train)
```

```python
classfier.score(x_test, y_test)
```
